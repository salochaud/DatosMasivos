```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Tarea 2 - FIFA RDD") \
    .getOrCreate()

rdd_players = spark.sparkContext.textFile("/content/players_22.csv")

rdd_players.take(5)
```

```python
header = rdd_players.first()
print(header)
```

```python
cols = header.split(",")
overall_idx = cols.index("overall")
potential_idx = cols.index("potential")
age_idx = cols.index("age")
pace_idx = cols.index("pace")
shooting_idx = cols.index("shooting")
passing_idx = cols.index("passing")
dribbling_idx = cols.index("dribbling")

data_rdd = rdd_players.filter(lambda line: line != header)
```

```python
overall_rdd = data_rdd.map(lambda line: float(line.split(",")[overall_idx]))

promedio_overall = overall_rdd.mean()
print(f"El promedio de overall de los jugadores en FIFA es: {promedio_overall:.2f}")
```

```python
potential_rdd = data_rdd.map(lambda line: float(line.split(",")[potential_idx]))

promedio_potential = potential_rdd.mean()
print(f"El promedio de potencial de los jugadores en FIFA es: {promedio_potential:.2f}")
```

```python
overall_sum = overall_rdd.sum()
potential_sum = potential_rdd.sum()
total = data_rdd.count()

brecha_promedio = (potential_sum - overall_sum) / total
print(f"La brecha promedio entre potencial y overall en FIFA es: {brecha_promedio:.2f}")
```

```python
pace_rdd = data_rdd.map(lambda line: float(line.split(",")[pace_idx]))
shooting_rdd = data_rdd.map(lambda line: float(line.split(",")[shooting_idx]))
passing_rdd = data_rdd.map(lambda line: float(line.split(",")[passing_idx]))
dribbling_rdd = data_rdd.map(lambda line: float(line.split(",")[dribbling_idx]))

print(f"Promedio de ritmo (pace):       {pace_rdd.mean():.2f}")
print(f"Promedio de disparo (shooting): {shooting_rdd.mean():.2f}")
print(f"Promedio de pase (passing):     {passing_rdd.mean():.2f}")
print(f"Promedio de regate (dribbling): {dribbling_rdd.mean():.2f}")
```

```python
elite_rdd = overall_rdd.filter(lambda x: x >= 85)
total_elite = elite_rdd.count()
porcentaje_elite = (total_elite / total) * 100

print(f"Jugadores con overall >= 85: {total_elite}")
print(f"Porcentaje de jugadores élite en el dataset: {porcentaje_elite:.2f}%")
```
